```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# liste des URLs des articles
urls = ["https://lematin.ma/express/2022/factures-energetique-alimentaire-seront-salees-maroc-entre-2022-2024/375134.html",
        "https://lematin.ma/express/2022/matieres-premieres-detente-prix-mois-flambee/384853.html",
        "https://lematin.ma/journal/2021/prix-mondiaux-gaz-butane-bondissent-58-fin-aout/364782.html",
        "https://lematin.ma/journal/2021/hausse-cours-mondiaux-parler-super-cycle/361460.html",
        "https://lematin.ma/journal/2021/dolidol-investit-110-millions-dh-apportes-berd-usine-recyclage/364323.html",
        "https://lematin.ma/express/2023/recettes-douanieres-hausse-82-fin-mars-2023/388752.html",
        "https://lematin.ma/express/2023/emballages-amg-packaging-soffre-usine-marrakech/388759.html",
        "https://lematin.ma/express/2023/comptes-desherence-mise-garde-cdg/388751.html",
        "https://lematin.ma/express/2023/fckm-finance-etudes-residents-medecine-pharmacie/388254.html",
        "https://lematin.ma/express/2023/vie-publique-citoyens-encadree-lois-caduques-ouahbi/388763.html"]

# initialiser une liste vide pour stocker les données des articles
articles = []

# initialiser le navigateur web
driver = webdriver.Chrome('C:\chromedriver\chromedriver')

# parcourir les URLs des articles
for url in urls:
    # ouvrir l'URL de l'article
    driver.get(url)
    
    # récupérer le titre de l'article
    titre = driver.find_element(By.ID, "title").text
    
    # récupérer la date de l'article
    date = driver.find_element(By.TAG_NAME, "time").text
    
    # récupérer le contenu de l'article
    contenu = driver.find_element(By.XPATH, "//article[@class='card p-1 mb-2']")
    
    # récupérer l'URL de l'article
    url_article = driver.current_url
    
    # ajouter les données de l'article à la liste des articles
    articles.append({'Titre': titre, 'Date': date, 'Contenu': contenu.text, 'URL': url_article})
    
# fermer le navigateur web
driver.quit()

# convertir la liste des articles en un DataFrame pandas
df = pd.DataFrame(articles)

```
